```python
import numpy as np
import matplotlib.pyplot as plt


g = np.array([
    0, 
    1,
    7/6,
    3/2,
    5/3,
    np.inf
    ])

b_func = lambda g: (g - 2) / (3*g - 4)
b = b_func(g)
b[-1] = 1/3 #((inf - 2) / (3 inf - 1))
```

```python
N = 500

M = np.ones((len(g), N)) * np.linspace(0, 1, N)[None, :]
R = M**b[:, None]
R[R>1e5] = np.nan
norm = 1/np.nanmax(R, axis=1)
R = R *norm[:, None]
```

```python

fig, ax = plt.subplots(1,2, figsize=(20, 10))


for i, beta in enumerate(b):
    ax[0].plot(R[i], M[i],label="$\\gamma = %.3f, \, \\beta = %.3f$"%(g[i], beta))

ax[0].legend()

gs = np.linspace(0, 3, 200)
bs = b_func(gs)
i = np.argmax(bs) + 1
print(i)
ax[1].plot(gs[:i ], bs[:i], "b")
ax[1].plot(gs[i:], bs[i:], "b")

ax[1].plot(g, b, "kx")
ax[1].plot(2.5, 0, "kx")
ax[1].annotate("", xy=(2.5 + 0.4, 0), xytext=(2.5+0.05, 0), arrowprops=dict(arrowstyle="->"))
print(g)

ax[1].set_ylim(-2, 2)

plt.show()
```
